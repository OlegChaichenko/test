```python
import pandas as pd
from io import BytesIO
from PIL import Image
import requests
from concurrent.futures import ThreadPoolExecutor
```

```python
id='1QX2IhFyYmGDFMvovw2WFz3wAT4piAZ_8hi5Lzp7LjV0'
name='data'
gsheet_url = "https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(id, name)


df=pd.read_csv(gsheet_url)

list=[]
for i in df['image_url']:
    list.append(i)

```

```python
def get_image_dimensions(image_url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
        response = requests.get(image_url,headers=headers )
        if response.status_code == 200:
            image = Image.open(BytesIO(response.content))
            width, height = image.size
            print(width,height)
            return width,height
            #time.sleep(1)
        else:
            print("Помилка отримання відповіді:", response.status_code)
            return print("Помилка отримання відповіді:", response.status_code)
    except Exception as e:
        print("Виникла помилка:", str(e))
```

```python
list2=[]
list_test=[]
for i in range(0,len(list)):
    list_test.append(list[i])

print(len(list_test))
with ThreadPoolExecutor(max_workers=25) as executor:
    for dimensions in executor.map(get_image_dimensions, list_test):
        list2.append(dimensions)
print(list2)
print(len(list2))
tuple_strings = [str(t) for t in list2]

# Создаем новый DataFrame, объединяя столбец image_url и список кортежей
new_df = pd.DataFrame({'image_url': df['image_url'], 'size': tuple_strings})
print(new_df)
new_df.to_csv('new_file.csv', index=False)
```
